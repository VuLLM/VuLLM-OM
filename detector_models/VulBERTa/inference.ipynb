```python
from transformers import pipeline
import pandas as pd
from transformers import AutoTokenizer
```

```python
pipe = pipeline("text-classification", model="claudios/VulBERTa-MLP-Devign", trust_remote_code=True, device=0, return_all_scores=True)
tokenizer = AutoTokenizer.from_pretrained("claudios/VulBERTa-MLP-Devign", trust_remote_code=True)
```

```python
df = pd.read_csv("vulgen/only_vul/vulgen_only_vul.csv")
print(len(df))
```

```python

count = 0
correct = 0
for index, row in df.iterrows():
    func_value = row['func']
    tokens = tokenizer(func_value, return_tensors='pt')
    if tokens['input_ids'].shape[1] > 1026:
        continue
    x = pipe(func_value)
    count += 1
    if x[0][1]['score'] > 0.5:
        correct += 1
print(f"Correct: {correct}/{count} ({correct/count})")
```
