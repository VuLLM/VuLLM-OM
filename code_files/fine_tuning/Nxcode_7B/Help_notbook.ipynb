```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
tokenizer = AutoTokenizer.from_pretrained("NTQAI/Nxcode-CQ-7B-orpo", trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained("NTQAI/Nxcode-CQ-7B-orpo", trust_remote_code=True, torch_dtype=torch.bfloat16).cuda()
```

```python
input_text = "#write a quick sort algorithm"
inputs = tokenizer(input_text, return_tensors="pt", add_special_tokens=True, return_token_type_ids=False).to("cuda")
outputs = model.generate(**inputs, max_length=256, eos_token_id=tokenizer.eos_token_id)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
